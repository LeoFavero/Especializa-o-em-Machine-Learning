```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro
from scipy.stats import norm
from scipy import stats
import statsmodels.api as sm
import os
```

### Teste de Normalidade

```python
x = np.random.randn(100)
stats, p = shapiro(x)
print(stats)
print(p) # se o p-valor for menor que a significancia alpha, a hipotese nula é rejeitada
```

### Teste de comparação da média de um conjunto de dados vs uma constante ou outra média

```python
x1 = norm.rvs(loc=10000, scale=1000, size=50)
x2 = norm.rvs(loc=1000, scale=100, size=50)
x3 = norm.rvs(loc=1000, scale=100, size=50)

print(x1.mean())
print(x2.mean())
print(x3.mean())
```

```python
stats.ttest_1samp(x1,10000)
```

```python
stats.ttest_1samp(x2,1000)
```

```python
stats.ttest_1samp(x3,1000)
```

```python
stats.ttest_ind(x1,x2)
```

```python
stats.ttest_ind(x1,x3)
```

```python
stats.ttest_ind(x2,x3)
```

## Exercício HBAT.csv

```python
os.chdir('/Users/leonardofaverobocardi/Documents/GitHub/Python_Bootcamp/Módulo 02/db')
hbat = pd.read_csv('HBAT.csv')
hbat.head(3)
```

```python
arr = ['x9','x10','x11','x12','x13','x14','x15','x16','x17','x18']

for val in arr:
    
    x = pd.DataFrame(hbat[val]) #independent
    y = pd.DataFrame(hbat['x19']) #dependent

    x = sm.add_constant(x)
    model = sm.OLS(y,x).fit()
    predictions = model.predict(x) #faz as predições do modelo

    print(model.summary())
    print('\n')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x = ['x' + str(i) for i in range(9,19)]
x = hbat[x]
y = hbat['x19']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=0)

linearRegressor = LinearRegression()
linearRegressor.fit(x_train, y_train)

print('intercept:',linearRegressor.intercept_) #intercepto ou coef linear
print('coef:',linearRegressor.coef_) #coef angular ou slope

#previsao
y_pred = linearRegressor.predict(x_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
from sklearn import metrics

#MAE
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred))

#MSE
print('Mean Squared Error:',metrics.mean_squared_error(y_test, y_pred))

#RMSE
print('Root Mean Squared Error:',np.sqrt(metrics.mean_absolute_error(y_test, y_pred)))
```
