## Excercício 3

### Considere um arquivo CSV (comma separated values) com os seguintes com dados de todas as copas do mundo até o ano de 2014 (WorldCups.csv):
- Year - Ano da copa
- Country - Pais sede
- Winner - Campeão
- Runners-Up - Vice-Campeão
- Third - Terceiro lugar
- Fourth - Quarto Lugar
- GoalsScored - Total de Gols
- QualifiedTeams - Quantidade de Times
- MatchesPlayed - Quantidade de Jogos
- Attendance - Publico Total

### Pede-se
* Crie uma função que efetue a leitura do arquivo e crie um arquivo de saida com as seguintes informações (uma informação por linha):
    - Soma de público das copas com anos final 0 (1930, 1950, etc)
    - Quantidade total de gols entre as copas de 1954 e 1990, inclusive
    - Média de público
    - Média de gols por partida
    - Quantidade de vezes em que o país sede foi campeão
    - Quantidade de vezes em que o time do Brasil ficou entre uma das 4 primeiras posições
    - Ano das edições em que o time da França finalizou em terceiro lugar
    - Quantidade de vitórias por país, classificada em ordem crescente do número de títulos


```python
import csv

def world_cup(file):
    
    # creating variables
    attendance_total = 0
    goals_total = 0
    attendance_all = 0
    editions = 0
    attendance_avg = 0
    goals_all = 0
    matches = 0
    goals_avg = 0
    host_champ = 0
    brazil_top = 0
    france_third = []
    country_win = {}
    
    
    # open file with csv library
    with open(file) as csv_file:
        csv_reader = csv.DictReader(csv_file)
        # could use csv.reader(csv_file) to use list index instead
        
        # iterate through source file
        for row in csv_reader:
            
            # removes the . from the attendance values
            for arr in row.values():
                if '.' in arr:
                    row['Attendance'] = arr.replace('.','')
            
            # check for year ending in 0 -> sum the attendance
            if int(row['Year'][3]) == 0:
                attendance_total += int(row['Attendance'])
                
            # check for years between -including- 1954 and 1990 -> sum the goals
            if int(row['Year']) >= 1954 and int(row['Year']) <= 1990:
                goals_total += int(row['GoalsScored'])
                
            # check for host country -> sum if host is champ
            if row['Country'] == row['Winner']:
                host_champ += 1
                
            # check for Brazil in top four -> sum the number of times it happpens
            if row['Winner'] == 'Brazil' or row['Runners-Up'] == 'Brazil' or row['Third'] == 'Brazil' or row['Fourth'] == 'Brazil':
                brazil_top += 1
            
            # check for year that France is Third -> append respective year on list
            if row['Third'] == 'France':
                france_third.append(row['Year'])
                
            # add the winning team to the dictionary key and count number of titles as values
            if row['Winner'] in country_win.keys():
                country_win[row['Winner']] += 1
            else:
                country_win[row['Winner']] = 1
            
            # ----- sum variables for future calculations -------
            
            # sum the attendance of all editions
            attendance_all += int(row['Attendance'])
            
            # sum the number of editions
            editions += 1
            
            # sum the goals of all editions
            goals_all += int(row['GoalsScored'])
            
            # sum the number of matches played
            matches += int(row['MatchesPlayed'])
                
        # calculate the attendance average of all editions
        attendance_avg = attendance_all / editions
        
        # calculate the average of goals of all editions
        goals_avg = goals_all / matches
        
        # sort the dictionary with wins by country
        country_win_sorted = {k: v for k, v in sorted(country_win.items(), key=lambda x: x[1])}
                      
                      
        # writing on new .txt file
        with open('WorldCupsInfo.txt', 'w') as txtFile:
        
            txtFile.write('- Soma de público das copas com final 0: {}\n'.format(attendance_total))      
            txtFile.write('- Quantidade total de gols entre as copas de 1954 e 1990: {}\n'.format(goals_total))
            txtFile.write('- Média de público: {}\n'.format(attendance_avg))
            txtFile.write('- Média de gols por partida: {0:.2f}\n'.format(goals_avg))
            txtFile.write('- Quantidade de vezes em que o país sede foi campeão: {}\n'.format(host_champ))
            txtFile.write('- Quantidade de vezes em que o time do Brasil ficou entre uma das 4 primeiras posições: {}\n'.format(brazil_top))
            txtFile.write('- Ano das edições em que o time da França finalizou em terceiro lugar: {} e {}\n'.format(france_third[0],france_third[1]))
            txtFile.write('- Quantidade de vitórias por país, classificada em ordem crescente do número de títulos:\n')
            for k,v in country_win_sorted.items():
                txtFile.write('\t{0:>10}: {1:>1} título(s)\n'.format(k,v))
```

```python
world_cup('WorldCups.csv')
```
