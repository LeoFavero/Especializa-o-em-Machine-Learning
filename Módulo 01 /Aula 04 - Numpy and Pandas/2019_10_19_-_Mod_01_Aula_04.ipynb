```python
df = pd.DataFrame({'A':[np.nan,2,3],'B':[np.nan,4,5],'C':[np.nan,7,np.nan]})
df
```

```python
df.dropna()
```

```python
df.dropna(how = 'all')
```

```python
df.fillna(0)
```

```python
df.fillna(df.min())
```

```python
df.fillna(method = 'ffill') # preenche com o ultimo dado valido daquela coluna
```

```python
df.fillna(method = 'bfill') # preenche com o dado de baixo da coluna
```

```python
df.fillna(value={'A': 0,'B': 10, 'C': 100})
```
